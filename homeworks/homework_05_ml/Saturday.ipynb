### Step 3. Assign it to a variable called crime.

```python
crime_rates = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv', sep=',')
crime_rates.head()
```

### Step 4. What is the type of the columns?

```python
crime_rates.dtypes
```

##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime_rates['Year'] = pd.to_datetime({'year':crime_rates['Year'], 'month':12, 'day':31},  format='%Y%m%d')
#crime_rates['Year'] = crime_rates['Year'].apply(lambda x: x.year)
crime_rates.dtypes
```

### Step 6. Set the Year column as the index of the dataframe

```python
crime_rates.set_index("Year", inplace=True)
crime_rates.head()
```

### Step 7. Delete the Total column

```python
crime_rates.drop(['Total'], axis=1, inplace=True)
crime_rates.head()
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
df = crime_rates.groupby(list(map(int, list(crime_rates.index.year/10))))[list(crime_rates.columns)[1:]].sum()
df
```

### Step 9. What is the mos dangerous decade to live in the US?

```python
df['Total'] = df.sum(axis=1)
print('The most dangerous in total crimes is {} decade'.format(df['Total'].idxmax()))
df
```
