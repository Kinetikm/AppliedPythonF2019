```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("adult.data", names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 
                                      'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 
                                     'hours-per-week', 'native-country', 'salary'])
```

```python
df.sex.value_counts()
```

```python
df.loc[df['sex'] == ' Male'].age.mean()
```

```python
len(df.loc[df['native-country'] == ' United-States']) / len(df['native-country'])
```

```python
plt.figure(figsize=(20, 10))
plt.ylabel('Number of people', fontsize=14)
plt.xlabel('Level of education', fontsize=14)
edu = df.groupby('education').size()
edu.plot.bar()
```

```python
df.groupby('education').size()
```

```python
num_of_people = df[df['salary'] == ' >50K'].groupby('education').size()
plt.figure(figsize=(20, 10))
plt.xlabel('Level of education', fontsize=14)
plt.ylabel('Number of people with salary >50K', fontsize=14)
num_of_people.plot.bar()
#      
```
