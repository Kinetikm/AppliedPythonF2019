```python
import pandas as pd
import csv
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('./Train.csv', index_col=0)
```

```python
df.head()
```

```python
# df.insert(1, 'year', pd.to_datetime(df['date']).dt.year)
# df.insert(2, 'month', pd.to_datetime(df['date']).dt.month)
```

```python
df.drop(columns=['date', 'street_id','n_photos', 'kw1', 'kw2', 'kw3','kw4','kw5','kw6','kw7','kw8', 'kw9','kw10','kw11','kw12','kw13'], inplace=True)
df.head()
```

```python
#date_col = []
#for i in range(len(df['year'])):
    #date_col += [df['year'][i] + df['month'][i]/10]
# df.insert(1, 'date', date_col)
```

```python
df.fillna(df.median(), inplace=True)
df.head()
```

```python
# df.drop(columns=['year', 'month'], inplace=True)
```

```python
x = df.iloc[:,:-1]
```

```python
x.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=4)
```

```python
y = df.iloc[:,-1]
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=3)
```

```python
x_poly = pf.fit_transform(x_train)
x_poly
```

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
lr = LinearRegression()
lasso = Lasso(alpha=1)
ridge = Ridge(alpha=1)
```

```python
lr.fit(x_poly, y_train)
```

```python
lasso.fit(x_poly, y_train)
```

```python
ridge.fit(x_poly, y_train)
```

```python
lasso.coef_
```

```python
ridge.coef_
```

```python
lr.coef_
```

```python
lr.score(pf.fit_transform(x_test), y_test)
```

```python
lasso.score(pf.fit_transform(x_test), y_test)
```

```python
ridge.score(pf.fit_transform(x_test), y_test)
```

```python
ridge.score(pf.fit_transform(x_train), y_train)
```
